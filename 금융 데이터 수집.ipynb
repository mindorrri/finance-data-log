# [프로젝트] 코스피와 코스닥 전종목 주가 데이터 수집하기

```python
import FinanceDataReader as fdr
```

```python
# 종목 목록 가져오기
stock_list = fdr.StockListing("KRX")
display(stock_list)
```

```python
#코넥스 종목 제외
stock_list = stock_list.loc[stock_list["Market"] != "KONEX"]
```

```python
print(stock_list.columns)
```

```python
# 종목 정보 내보내기
import os

save_dir = "../../데이터"
os.makedirs(save_dir, exist_ok = True)
save_path = os.path.join(save_dir, "종목정보.txt")
stock_list.to_csv(save_path, sep = "\t", index = False , encoding = "euc-kr")
```

```python
import time
for code, name in stock_list[['Code' , "Name"]].values:
    print(code, name)
    while True:
        try:
            data = fdr.DataReader (code, "2025-01-01", "2025-06-27")
            if len(data) > 300:
                data.to_csv("../../데이터/주가데이터/{}.csv".format(name))
            time.sleep(1)
            break
        except:
            time.sleep( 10 * 60)
```
